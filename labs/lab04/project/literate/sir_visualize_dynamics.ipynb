# Введение

Данный скрипт предназначен для итоговой визуализации результатов
параметрического сканирования коэффициента заразности β.

## Назначение

-   Загрузка результатов из CSV-файла
-   Построение составного графика с тремя панелями
-   Подготовка графиков для отчёта

# Инициализация проекта

```julia
using DrWatson
@quickactivate "project"
using Agents, DataFrames, Plots, CSV
include(srcdir("sir_model.jl"))
```

# Загрузка данных

```julia
# Загружаем результаты сканирования
df = CSV.read(datadir("beta_scan_all.csv"), DataFrame)
```

# Построение составного графика

```julia
# Создаём составной график
p1 = plot(df.beta, df.peak, label = "Пик", xlabel = "β", ylabel = "Доля инфицированных")
plot!(p1, df.beta, df.final_inf, label = "Конечная")

p2 = plot(df.beta, df.deaths, xlabel = "β", ylabel = "Число умерших")

p3 = plot(df.beta, df.final_rec, xlabel = "β", ylabel = "Доля выздоровевших")

plot(p1, p2, p3, layout = (3, 1), size = (800, 900))
savefig(plotsdir("comprehensive_analysis.png"))
```

# Описание панелей графика

## Панель 1: Пик и конечная доля инфицированных

Показывает зависимость максимальной и итоговой доли переболевших от
коэффициента заразности.

## Панель 2: Число умерших

Демонстрирует нелинейное увеличение смертности с ростом β.

## Панель 3: Доля выздоровевших

Отражает насыщение доли переболевших при высоких значениях β.

# Выводы

Сводный график позволяет одновременно оценить:

1.  **Порог возникновения эпидемии** (рост пика при β \> 0.3)
2.  **Нелинейное увеличение смертности**
3.  **Насыщение доли переболевших** при высоких β
